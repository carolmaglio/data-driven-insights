### **5. Estruturas Condicionais e Laços**

```python
# Escolher coluna numérica 'PRICEEACH'
preco_cada = 'PRICEEACH'
primeiro_valor = df[preco_cada][0]

if primeiro_valor > 50:
    print(f"{primeiro_valor} - Valor do preço unitário alto")
elif primeiro_valor > 68.86:
    print(f"{primeiro_valor} - Valor do preço unitário médio")
else:
    print(f"{primeiro_valor} - Valor do preço unitário baixo")

# Usar o laço 'for' para calcular os primeiros 5 valores da coluna
soma = 0

lista = list(df[preco_cada][:5])
for n in lista:
    soma += n
print("Soma dos 5 primeiros valores:", soma)

# While: Encontrar o primeiro valor maior que um certo limite, definido de acordo com os dados
limite = 50  # Ajuste conforme o contexto da coluna escolhida
i = 0

while i < len(lista) and lista[i] <= limite:
    i += 1
if i < len(lista):
    print("Primeiro valor >", limite, "encontrado:", lista[i])
else:
    print("Nenhum valor maior que", limite, "encontrado nos primeiros 5.")
```

### **6. Operadores Aritméticos e Manipulação de Dados**

```python
# Verificar se a coluna 'PRICEEACH' existe e cria a coluna 'PRECO_DESCONTO'
if 'PRICEEACH' in df.columns:
    df['PRECO_DESCONTO'] = df['PRICEEACH'] * 0.9
    print(df[['PRICEEACH', 'PRECO_DESCONTO']].head(5))

# Calcular a diferença entre o preço unitário e o preço com desconto
if 'PRICEEACH' in df.columns and 'PRECO_COM_DESCONTO' in df.columns:
    df['DIFERENCA'] = df['PRICEEACH'] - df['PRECO_DESCONTO']
    print(df[['PRICEEACH', 'PRECO_DESCONTO', 'DIFERENCA']].head(5))
```

### **7. NumPy e Arrays Numéricos**

```python
# Criação de um array NumPy a partir da coluna 'SALES'
array_sales = np.array(df['SALES'])

# Operações NumPy
total_sales = array_sales.sum()
media_sales = array_sales.mean()
max_sales = array_sales.max()
min_sales = array_sales.min()

# Exibição dos resultados com formtados em reais
print(f"Slicing: {array_sales[2:5]}")
print(f"Total de vendas: {total_sales:.2f}")
print(f"Média de vendas: {media_sales:.2f}")
print(f"Maior venda: {max_sales:.2f}")
print(f"Menor venda: {min_sales:.2f}")
```

### **8. Acesso e Manipulação de Dados com Pandas**

```python
print("Receita por categoria:")

# Agrupar os dados por categoria de produto e realizar a soma do total de vendas por categoria
sales_productline = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)
print(sales_productline)
```

### **9. Visualização de Dados**


### **9.1 Gráfico de Linhas**
*   Com o gráfico de linhas é possível visualizar a receita gerada por cada categoria de produto.
*   A categoria de **maior receita** é "Classic Cars", destacando-se por ser o produto mais vendido.
*   A categoria de **menor receita** é "Trains", apresentando o menor valor de vendas.





```python
if 'PRODUCTLINE' in df.columns and 'SALES' in df.columns:
    # Agrupar por 'PRODUCTLINE' e somar as vendas, depois ordenar do menor para o maior
    vendas_por_categoria = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=True)

    # Criar o gráfico de linha com marcadores
    ax = vendas_por_categoria.plot(kind='line', marker='o')

    # Adicionar rótulos de dados em cada ponto
    for i, (categoria, valor) in enumerate(vendas_por_categoria.items()):
        ax.text(i, valor, f'{valor:,.0f}', ha='center', va='bottom', fontsize=7)

    # Personalizar o gráfico
    plt.title('Receita por Categoria')
    plt.xlabel('Categoria')
    plt.ylabel('Receita Total')
    plt.xticks(rotation=45)

    # Exibir o gráfico
    plt.show()
```

### **9.2 Gráfico de Barras**

*   Com o gráfico de barras, é possível comparar a quantidade de vendas por país.
*  O país com a maior venda foi **"EUA"**.
*  O país com a menor venda foi **"Ireland"**.




```python
if 'COUNTRY' in df.columns and 'SALES' in df.columns:
    # Agrupar e somar as vendas por país
    vendas_por_pais = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False)

    # Criar o gráfico de barras
    ax = vendas_por_pais.plot(kind='bar', figsize=(10, 6))

    # Adicionar rótulo de dados em cada barra com rotação vertical (90°)
    for i, valor in enumerate(vendas_por_pais):
        ax.text(i, valor, f'{valor:,.0f}', ha='center', va='bottom', fontsize=6, rotation=90)

    # Personalizar o gráfico
    plt.title('Venda por País')
    plt.xlabel('País')
    plt.ylabel('Receita Total')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.5)

    # Exibir o gráfico
    plt.show()
```

### **9.3 Gráfico de Dispersão**

*   O gráfico de dispersão nos mostra se há relação entre a quantidade de receitas e a quantidade de itens encomendados.

```python
if 'QUANTITYORDERED' in df.columns and 'SALES' in df.columns:
    plt.scatter(df['QUANTITYORDERED'], df['SALES'])
    plt.title('Dispersão: Quantidade Encomendada vs Receita')
    plt.xlabel('Quantidade encomendada')
    plt.ylabel('Receita Total')
    plt.show()
```
